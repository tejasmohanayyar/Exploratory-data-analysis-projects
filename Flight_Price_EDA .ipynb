```python
fprice_train = pd.read_excel('data/Flight_Price/Data_Train.xlsx',parse_dates = ['Date_of_Journey'])
fprice_test = pd.read_excel('data/Flight_Price/Test_set.xlsx',parse_dates = ['Date_of_Journey'])
```

```python
def datamanipulation(fprice_train,inplace = False):
    fprice_train['Dep_Time'] = pd.to_datetime(fprice_train['Dep_Time'])
    fprice_train['Arrival_Time'] = pd.to_datetime(fprice_train['Arrival_Time'])

    duration_hour = []
    duration_min = []

    for i in fprice_train['Duration']:
        x = i.split('h')
        try:
            duration_hour.append(int(x[0]))
        except:
            duration_hour.append(0)
        try:
            duration_min.append(int(x[1].split(' ')[1].split('m')[0]))
        except:
            duration_min.append(0)
            
    fprice_train['Date Day'] = fprice_train['Date_of_Journey'].dt.day
    fprice_train['Date Month'] = fprice_train['Date_of_Journey'].dt.month
    
    fprice_train['Duration_hour'] = duration_hour
    fprice_train['Duration_min'] = duration_min
    fprice_train['Arrival hour'] = fprice_train['Arrival_Time'].dt.hour
    fprice_train['Arrival min'] = fprice_train['Arrival_Time'].dt.minute
    fprice_train['Dep hour'] = fprice_train['Dep_Time'].dt.hour
    fprice_train['Dep min'] = fprice_train['Dep_Time'].dt.minute
    
    fprice_train.drop(['Dep_Time','Arrival_Time','Duration','Date_of_Journey'],axis = 1, inplace = inplace)
    return fprice_train.head()
```

```python
datamanipulation(fprice_train,inplace = True)
```

```python
datamanipulation(fprice_test,inplace = True)
```
